```python
%matplotlib inline
# 노트북에서 보이기
import pandas as pd
import xlwings as xw
# 네모 해결
from matplotlib import rc
rc('font', family="Malgun Gothic")
```

```python
df = pd.read_excel('score.xlsx', index_col='구분')
```

```python
# 기준 정하기
매출_top10  = df.sort_values('매출', ascending=False).head(5)
매출_top10
```

```python
# 그래프 그리기
ax = 매출_top10.plot(kind='barh', grid=True, figsize=(8,3))
```

```python
from matplotlib import font_manager
font_manager.fontManager.ttflist
```

```python
매출_top10[['개정안']].sort_values('개정안').plot(kind='bar')
```

```python
# 파이그래프
매출_top10[['개정안','매출']].sort_values('개정안').plot(kind='pie', y='매출')
```

```python
# 차트 이미지 엑셀에 추가하기
xw.view(df)
```

```python
fig = ax.get_figure()
fig
```
